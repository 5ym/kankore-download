```
from google.colab import files
import requests
import zipfile
# urlからファイルをダウンロード
def download_file(url):
    filename = url.split('/')[-1]
    r = requests.get(url, stream=True)
    if r.status_code != requests.codes.ok:
      return False
    with open(filename, 'wb') as f:
        for chunk in r.iter_content(chunk_size=1024):
            if chunk:
                f.write(chunk)
                f.flush()
        return filename
```

```
#タイトルコール1
z = zipfile.ZipFile("title1.zip", "w")
for i in range(1,1000):
  number_padded = '%03d' % i
  filename = download_file('http://125.6.189.7/kcs2/resources/voice/titlecall_1/'+number_padded+'.mp3')
  print(i)
  if filename == False:
    print('file get complete')
    break;
  z.write(filename)
z.close()
files.download('title1.zip')
```
